<div style="text-align: justify;">
Мембранным потенциалом называют разность потенциалов, возникающюю на мембране клетки, причём разность принято считать между внутренней и наружной сторонами мембраны. Харовые водоросли, как и многие другие эукариотические организмы, обладают мембранным потенциалом. Он необходим клеткам водоросли для поглощения питательных веществ из окружающей среды, а также для транспорта различных веществ через плазматическую мембрану. Основным элементом, учавствующим в создании мембрранного потенциала у харовых водорослей является H-ATPase. 
<div>

```python
from IPython.display import Image
Image(filename='ATPase.png')
```

<div style="text-align: justify;">
Работа H-ATPase, выполняющей роль протонной помпы, приводит к выбросу значительного количества протонов изнутри клетки наружу, закислению окружающей клетку прослойки наружной среды, а также созданию электрохимического градиента на плазматической мембране. Именно активная работа H-ATPase вносит решающий вклад в создание мембранного потенциала харовых водорослей (смотрите источник [2]), который устанавливается на уровне -200mV, часто принимая ещё более отрицательные значения. Такие большие значения мембранного потенциала (здесь и далее величина мембранного потенциала имеется ввиду по модулю) приводят к отличным от таковых наблюдаемых в животных клетках движениям ионов. Здесь стоит отметить, что то, как ионы будут двигаться через мембрану, определяется их двтжущей силой, которую можно определить как разность их равновесного потенциала и мембранного потенциала. Расчёт равновесного потенциала в первом приближении можнор прооводить при помощи уравнения нернста, согласно которому $$V_{eq} = \frac{RT}{zF} \ln\left(\frac{C_{out}}{C_{in}}\right)$$ В таком случае движущая сила для иона выражается как $$\text{Driving force} = V_{eq} - V_{m}$$ Теперь рассмотрим, как будет направлен ток калия при потенциале покоя (потенциалом покоя называют мембранный потенциал клетки в состояния покоя). Для ионов калия существует градиент концентраций, при котороим внутри клетки концентрация ионов составляет около 100mM, а в пресной воде, окружающей клетку, концентрация калия мала и близка к 0.1mM. Если расчитать значение равновесного потенциала для ионов калия по представленному выше уравнению Нерста, приняв температуру окружающей среды (а, соответственно, и клетки) за  20°C, то получим значение -174mV. Таким оюразом, при значениях мембранного потенциала, соответсвующих потенциалу покоя харовой водоросли, будет происходить не выход калия из клетки, как наблюдается в случае животных клеток, а, наоборот, вход калия внутрь клетки. Тем не менее, ккак и в случае животных клеток, концентрации натрия во внеклеточной среде выше, чем во внутриклеточной, однако в случае растительных клеток, а также клеток харовых водорослей, ионы натрия не играют такой значительной роли и присутствуют в очень небольших концентрациях (около 1mM в пресной воде и ещё меньше внутри клеток). Здесь также стоит отметить, что ионы кальция в большей концентрации присутствуют снаружи клетки, что ишграет важную роль в процессах сигналинга и электрической возбудимости. Кроме того, существуют внутриклеточные запасы кальция, находящиеся внутри эндоплазматичемского ретикулума и способные его покидать в ответ на определённые стимулы.

__Электрические характеристики мембраны клетки__

<div style="text-align: justify;">
Как уже было сказано выше, плазматическая мембрана харовых водорослей обладает некоторым потенциалом покоя, значения которого достигают -200mV и больше. Кроме того, как и большинство мембран других клетки,  плазматическая мембрана харовых водорослей несёт в себе множество каналов. Каналами называют особые белки, способные служить своеобразной порой для ионов, проходящих через мебрану. Здесь стоит напомнить, что, как следует из уравнения Борна $$\Delta G = \frac{z^2 e^2}{8 \pi \epsilon_0 r} \left( \frac{1}{\epsilon_m} - \frac{1}{\epsilon_w} \right)
$$ потенциальная энергия перехода заряженного иона через неполяорную мембрану очень велтка, а это приводит к тому, что транспорт ионов через мембрану без каких-либо посредников оказывается практически невозможным. Каналы же способны помочь ионам в преодолении этого потенциального барьера, изменяя разницу в диэлектрической проницаемости мембранного бислоя и омывающих его водных растворов в сторону меньших значений. Кроме того, каналы на входе имеют специфический фильтр, способный сделать канал селективным для определённого типа ионов. Такой фильтр может селективно связать определённый тип ионов электростатическими взаимодействиями и лишить гидратной оболочки, препятствующей прохоэждению иона через мембрану. Таким образом, для перехода определённого типа ионов через мембрану нужны определённые же каналы, специфичные для одного или нескольких ионов. Разные участки мембраны обладают разной плотностью каналов на единицу площади мембраны, что влияет на проницаемость конкретного участка мембраны для данного типа ионов. Здесь стоит отметить, что плотность каналов на участке мембраны количественно описывают удельным сопротивлением мембраны $${R_m}$$ имеющим единицы измерения $$\text{Om} \cdot \text{cm}^2$$ тепрь стоит отметить, что уникальное строение мебрнаны в виде бислоя, несущего с двух сторон заряженные полярные головки фосфолипидов, а также сульфатированных гликолипидов, в то же время с довольно толстой (по сравнению с заряженными усчастками) прослойкой неполярных частей липилдов, позвооляет рассматривать мембрану в роли конденсатора, обкладками которого являются примембранные слои омывающих растворов по обе стороны мембраны. Конденсатор можно описать такой количественной характеристикой как ёмкость, причём в случае мембранной ёмкости представляется более удобным использовать удельную ёмкость $${C_m}$$ которую можно выразить в следующих единицах измерения $$ \frac{\text{μF}}{\text{cm}^2}$$

_Представление мембраны в виде R-C цепочки_

<div style="text-align: justify;">
Как было сказано выше, плазматическая мембрана обладает такими характеристиками, как удельное сопротивление и удельная ёмкость, а, значит, при прохожднении через неё электрического тока возможно наблюдать изменение мембранного потенциала. Поскольку мембрана клетки олбладает как каналами для ионов, которые характьеризуют сопротивление мебраны, так и ёмкостью, то можно представить плазматическую мембрану вы виде эквивалентной схемы, представленной на рисунке ниже (по источнику [3]):
<div>

```python
from IPython.display import Image
Image(filename='IMG_1483.jpg')
```

В таком случае, плотность тока, протекающего через мембранну, можно представить как сумму тока через сопротивление мембраны и емкостного тока: $$j_t = j_r + j_c$$


Ёмкость определяется как $$C = \frac{q}{\varphi_m}$$ а ток определяется как $$\frac{dq}{dt}$$ то емкостной ток можно представить как $$j_c = C_m \frac{d\varphi_m}{dt}$$ тогда $$j_t = \frac{\varphi_m}{R_m} + C_m \frac{d\varphi_m}{dt}$$ 

$$
\begin{aligned}
j_t R_m - \varphi_m &= R_m C_m \frac{d\varphi_m}{dt} \\
\int_{0}^{t} \frac{d\tilde{t}}{R_m C_m} &= \int_{0}^{\varphi_m} \frac{d\tilde{\varphi}_m}{j_t R_m - \tilde{\varphi}_m} \\
\frac{t}{R_m C_m} &= \ln \left( \frac{j_t R_m - \varphi_m}{j_t R_m} \right) \\
\frac{j_t R_m - \varphi_m}{j_t R_m} &= e^{-\frac{t}{R_m C_m}} \\
j_t R_m - \varphi_m &= j_t R_m e^{-\frac{t}{R_m C_m}} \\
\varphi_m &= j_t R_m \left( 1 - e^{-\frac{t}{R_m C_m}} \right)
\end{aligned}
$$


В итоге получаем уравнение для изменение потенциала при стимуляции мембраны током:

$$
\begin{aligned}
\varphi_m &= j_t R_m \left( 1 - e^{-\frac{t}{R_m C_m}} \right)
\end{aligned}
$$

Представленное выше урвавнение можно переписать в немного другом виде, если заменить произведение удельных сопротивления и ёмкости мембраны в показателе экспоненты на константу $$ \tau=R_mC_m$$

$$
\begin{aligned}
\varphi_m &= j_t R_m \left( 1 - e^{-\frac{t}{\tau}} \right)
\end{aligned}
$$

Данную величину также называют постоянной времени, поскольку она характеризует, насколько быстро происходит перезарядка мембраны до 0.36 нового значения мембранного потенциала.

Если же прекратить стимуляцию, то мембранный потенциал будет возвращаться к исходному значению согласно закону:

$$
\begin{aligned}
\varphi_m &= V_0e^{-\frac{t}{\tau}}
\end{aligned}
$$

где $$V_0$$ равновесное значение мембранного потенциала, которое достигается при бесконечно длительной стимуляции.

```python
from IPython.display import Image
Image(filename='IMG_1484.jpg')
```

Как следует из приведённых выше уравнений, такие электрические ответы мембраны подчиняются экспоненциальным законам при спаде и нарастании и называются пассивными электрическими свойствами мембраны. Однако мембрана харовых водорослей, как и в случае некоторых других водорослей и растений, спосбна к генерации потенциалов действия. Потенциалом действия называют электрический ответ мембраны, в ходе которого происходит резкое изменение значения потенциала от потенциала покоя амплитудой около 100mV, причём данное изменение потенциала мембраны способно распространяться на значительное растояние по мембране клетки без ощутимого затухания, в то время как пассивные характеристики мембраны такими свойствами не обладают.

_Потенциал действия в растительных клетках и клетках водорослей_

<div style="text-align: justify;">
Как уже было сказано выше, основным потенциал образующим ионом в растительной (или водорослёвой) клетке является протон. Кроме того, ионы калия представлены вы клетке в довольно значительной концентрации (100mM). Из этого всего следует, что потенциал действия растительной (водорослёвой) клетки будет значительно отличаться от такового в животной клетке. Действительно, как ионы, образующие потенциал действия, так и фазы его будут отличны от животных аналагов. Фазой ноль принято считать состояние мембраны до начала деполяризации. ак видно из рисунка ниже, на этой стаддии основной вклад вносит H-ATPase. Далее наступает фаза 1 быстрой деполяризации мембраны. Основную роль в этой фазе играют открывающинеся кальциевые каналы DmGLR3.6. Резкий вход кальция в клетку по этим каналам приводит к кальций зависимому выбросу дополнительных ионов кальция из эндоплазматического ретикулума, что также ускоряет деполяризацию. Кроме того, ускорение деполяризации в фазу 1 также обусловлено отккрытием кальций зависимых анионных каналов R-type DmQUAC1/ALMT, через которые происходтит выход ионов хлора из клетки. Далее следует фаза 2, в ходе которой начинается выход ионов калия по outward K+ channels и замедляется начальная быстрая деполяризация. Далее потенциал действия достигает своего пика, при значениях потенциала которого идёт уже значительный выход ионов калия из клетки. Далее наступает фаза 4 потенциала действия, в ходе которой происходит реполяоризация за счёт дальнейшего выхода калия из клетки через outward-rectifying (DmSKOR) K+ channels. Пик реполяризации достигается в фазу 5, причём было выяснено (смотрите источник 4), что H-ATPase ответственна за овершут реполяризации. Дальнейшему прохождению гиперполяризации препятсятвует активация потециал зависимых inward rectifying K+ channels, которые также вместе с хлорными каналами приводят к восстановлению ионного состава клетки. Такой потенциал действия характерен для венериной мухоловки, но он в целом консервативен и для харовых водорослей.

На графике отчётливо видно, что в ходже такого эксперимента на мембране возникали как пассивные электротонические потенциалы, так и потенциал действия. Стоит отметить, что амплитуда и форма электротоничексих ответов была значительно различной до ннаступления потпенциала действия, в ходе него и после его окончания. Как  видно из рисунка, до потенциала действия амплитуда электротонических потенциалов была максимальной, а их нарастание и спад происходили по эксмпоненциальному закону. В пике же потенцтиала действия амплитуда электротонических потенциалов достигла своегшо минимума, а характеры нарастания и спада этих потенциалов приняли прямоугольную форму. Далее с постепенным прохожденинем фаз потенциала действия, происходило постепенное восстановление амплитуды электротонических потенциалов, хотя поолного восстановления амплитуды к окончагнию потенциала действия не наблюдалось. Пассивные характеристики мембраны определяются её сопротивлениеим и ёмкостью, а потому изменение амплитуды данных электротонических потенциалов может быть объяснено по закону Ома. Из закона ома следует, что чем выше сопротивление проводника, тем большее смещение напряжения (в данноим случае мембранного потенциала) может быть достигнуто при неизменном подаваемом токе. Как было сказано в теоретическом введении, при возникновени потенциала действия значительно увеличивается проводимость мембраны за счёт открытия большого числа ионных каналов, а, значит, резко падает сопротивление мембраны. По закону Ома это должно привести к падению амплитуды электротонического потенциала, чьто и наблюдается в эксперименте.

_Исследование пассивных свойств мембраны_

<div style="text-align: justify;">
В ходе данного эксперимента было проведено измерение удельного сопротивления мембраны и удельной ёмкости мембраны. Для этого производили пять последовательных стимуляций клетки хары с амплитудой стимула 2.5V и длительностью 500ms. Далее производилось усреднение кривых для избавления от шума, так как шум можно описать случайным распределением, а матожидание случайных отклонений должно быть равно нулю. Полученная усреднённая кривая электротонического потенциала представлена на рисунке ниже:
<div>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data; header=None ensures the first row is kept as data
df = pd.read_csv('Chara_unknown.csv', header=None, names=['1', '2'])

# Set style for a clean, scientific look
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5))

# Plotting: Col '2' is Wavelength (X), Col '1' is Intensity (Y)
# sns.lineplot connects existing points without interpolation
ax = sns.lineplot(x='1', data=df, y='2', linewidth=1.5, color='darkblue')

# Scientific labeling
plt.xlabel('Time', fontsize=12, fontweight='bold')
plt.ylabel('Amplitude of electric response (mV)', fontsize=12, fontweight='bold')
plt.title('Chara electrotonic potential', fontsize=14)

plt.tight_layout()
plt.show()
```

<div style="text-align: justify;">
По полученной кривой определили максимальное отклонение мембранного потенциала, оно составило 85.386mV. Далее, зная амплитуду стимула в 2.5V и сопротивление нагрузочного резистора в 27MOm, определили, что сопротивление плазматической мембраны составляет 104205Om*cm^2. Далее произвели фиттирование экспериментальной кривой экспоненциальным уравнением и получили постоянную времени плазматической мембраны клетки хары: 154.1ms. Далее, зная значение удельного мембранного сопротивдлпения, а также значение постоянной времени, получили, что удельная ёмкость мембраны составляет 1.48μF/cm^2.
